## Categorical data plots

```python
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
```

```python
table=pd.crosstab(df_train.Gender,df_train.Response)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Gender VS Response')
plt.xlabel('Gender')
plt.ylabel('Response')
plt.savefig('Gender_VS_Response')
```
